```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from google.colab import files

# 1. MENCARI & LOAD DRY BEAN DATASET
print("Upload file DryBeanDataset.xlsx")
uploaded = files.upload()
df = pd.read_excel(list(uploaded.keys())[0])
print(f"Dataset loaded: {df.shape}")

# 2. PREPROCESSING: Normalisasi, Encoding, Train-Test Split
X = df.drop('Class', axis=1).values
y = LabelEncoder().fit_transform(df['Class'].values) # Encoding Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Split 70-30

mean = X_train.mean(axis=0)
std = X_train.std(axis=0) + 1e-8
X_train = (X_train - mean) / std # Normalisasi
X_test = (X_test - mean) / std
print(f"Data siap. Train: {X_train.shape}, Test: {X_test.shape}")

# 3. ARSITEKTUR ANN + 4. INIT BOBOT + AKTIVASI + FORWARD PROPAGATION
class ANN_Forward:
    def __init__(self, input_size, hidden_size, output_size):
        # INISIALISASI BOBOT UNIFORM sesuai spek
        self.w1 = np.random.uniform(-0.5, 0.5, (input_size, hidden_size))
        self.b1 = np.zeros((1, hidden_size))
        self.w2 = np.random.uniform(-0.5, 0.5, (hidden_size, output_size))
        self.b2 = np.zeros((1, output_size))
        print(f"\nArsitektur: {input_size} -> {hidden_size} -> {output_size}")
        print("Bobot diinisialisasi Uniform[-0.5, 0.5]")

    def tanh(self, x):
        # FUNGSI AKTIVASI TANH
        return np.tanh(x)

    def softmax(self, x):
        # FUNGSI AKTIVASI OUTPUT
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward_propagation(self, X):
        # FORWARD PROPAGATION DOANG
        print("\n=== FORWARD PROPAGATION ===")

        # Layer 1: Input -> Hidden
        z1 = X @ self.w1 + self.b1
        a1 = self.tanh(z1) # Aktivasi Tanh
        print(f"Hidden Layer output shape: {a1.shape}")

        # Layer 2: Hidden -> Output
        z2 = a1 @ self.w2 + self.b2
        a2 = self.softmax(z2) # Aktivasi Softmax
        print(f"Output Layer shape: {a2.shape}")
        print("Forward Propagation Selesai ✅")

        return a2

# JALANIN
input_size = X_train.shape[1] # 16 fitur Dry Bean
hidden_size = 64 # bebas, sesuai spek tugas
output_size = len(np.unique(y)) # 7 kelas Dry Bean

model = ANN_Forward(input_size, hidden_size, output_size)

# Coba 5 data pertama aja buat demo forward
output = model.forward_propagation(X_test[:5])
print("\nContoh Output Probabilitas 5 data pertama:")
print(output)
print("\nPrediksi kelas:", np.argmax(output, axis=1))
```
